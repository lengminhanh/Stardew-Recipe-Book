setting up the kitchen

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np  
import re
```

buying ingredients

```python
url = 'https://stardewvalleywiki.com/Foraging#Foraged_Items'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

print(soup)
```

df setup

```python
table = soup.find_all('table', class_ = 'wikitable')
table_basic = table[3]
table_spring = table[4]
table_summer = table[5]
table_fall = table[6]
table_winter = table[7]
table_beach = table[8]
table_mines = table[9]
table_desert = table[10]
table_gingerisland = table[11]
print(table_basic)
```

```python
titles = table_basic.find_all('th')
print(titles)
```

```python
ans = [title.text.strip() for title in titles]
print(ans)
```

```python
k = 0
table_basic_titles = []
for t in ans:
    if k < 3 :
        table_basic_titles.append(t)
    k+=1
print(table_basic_titles)    
```

```python
df = pd.DataFrame(columns = table_basic_titles)
df
```

plating on the table

```python
column_datas = table_spring.find_all('tr')

for row in column_datas[1:]:
   row_data = row.find_all('td')
   ans = [' '.join(data.stripped_strings) for data in row_data]
   
   if len(ans) > 4:
      row_image1 = row.find('img')
      res = row_image1['src']
      row_image = "https://stardewvalleywiki.com" + res

      table_spring_individual_row =[row_image]
      table_spring_individual_row.append(ans[1])
      table_spring_individual_row.append(ans[2])

      length = len(df)
      df.loc[length] = table_spring_individual_row
      print(table_spring_individual_row)
```
