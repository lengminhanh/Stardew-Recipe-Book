buying ingredients

```python
url = 'https://stardewvalleywiki.com/Cooking#Ingredient_Dishes'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
table = soup.find_all('table', class_ = 'wikitable')
table_stock = table[10]
print(table_stock)
```

```python
titles = table_stock.find_all('th')
print(titles)
```

```python
ans = [title.text.strip() for title in titles]
print(ans)
```

```python
k = 0
table_stock_titles = []
for t in ans:
    if k < 3 :
        table_stock_titles.append(t)
    k+=1
print(table_stock_titles)     
```

```python
df = pd.DataFrame(columns = table_stock_titles)
df
```

```python
column_datas = table_stock.find_all('tr')

for row in column_datas[1:]:
   row_data = row.find_all('td')
   ans = [' '.join(data.stripped_strings) for data in row_data]
   
   if len(ans) > 2:
      row_image1 = row.find('img')
      res = row_image1['src']
      row_image = "https://stardewvalleywiki.com" + res

      table_stock_individual_row =[row_image]
      length = len(df)
      print(table_stock_individual_row)
```
