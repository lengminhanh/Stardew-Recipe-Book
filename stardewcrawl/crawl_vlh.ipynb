```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np  
import re
```

```python
url = 'https://stardewvalleywiki.com/Fish'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

print(soup)
```

cooking plating on the table

```python
table = soup.find_all('table', class_ = 'wikitable')
table_pole = table[0]
table_crabpot = table[4]

print(table_pole)
print(table_crabpot)
```

```python
titles = table_pole.find_all('th')
print(titles)
```

```python
ans =[title.text.strip() for title in titles]
print(ans)
```

```python
k = 0
table_pole_titles = []
for t in ans:
    if k < 3 :
        table_pole_titles.append(t) 
    k+=1
print(table_pole_titles)
```

```python
df = pd.DataFrame(columns = table_pole_titles)
df
```

```python
column_datas = table_pole.find_all('tr')
```

```python

j = 0
for row in column_datas[1:]:
   if j== 13:
      j=0
   if j == 0:
      row_data = row.find_all('td')
      row_image1 = row.find('img')
      res = row_image1['src']
      row_image = "https://stardewvalleywiki.com" + res
      ans = [data.text.strip() for data in row_data]
      i = 0
      table_pole_individual_row =[row_image]
      for data in ans:
         if i < 3:
            if i!=0:
               table_pole_individual_row.append(data)
         else:
            break
         i+=1
      length = len(df)
      df.loc[length] = table_pole_individual_row
      print(table_pole_individual_row)
   
   j+=1
```
